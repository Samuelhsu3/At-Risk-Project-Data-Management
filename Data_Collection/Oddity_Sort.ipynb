# Sort Oddity Files (.sh)

This creates a new oddity session folder for every pair of raw files (should be edat2 and txt with the same name)in the current directory. 

```bash
#!/bin/bash

declare -A file_groups
sorted_file_count=0
sorted_folder_count=0
missing_files=()

for file in *.{txt,edat2}; do
    [ -e "$file" ] || continue
    base_name="${file%.*}"
    file_groups["$base_name"]+="${file##*.} "
done

for group in "${!file_groups[@]}"; do
    has_txt=false
    has_edat2=false
    directory_path="./$group"

    for ext in ${file_groups[$group]}; do
        [[ $ext == "txt" ]] && has_txt=true
        [[ $ext == "edat2" ]] && has_edat2=true
    done

    if $has_txt && $has_edat2; then
        mkdir -p "$directory_path"
        for ext in ${file_groups[$group]}; do
            mv "${group}.${ext}" "$directory_path"
            ((sorted_file_count++))
        done
        ((sorted_folder_count++))
    else
        [[ ! $has_txt ]] && missing_files+=("$group.txt")
        [[ ! $has_edat2 ]] && missing_files+=("$group.edat2")
    fi
done

if [ ${#missing_files[@]} -gt 0 ]; then
    echo "Missing Files:"
    for file in "${missing_files[@]}"; do
        echo "$file"
    done
fi

echo "Total Files Sorted: $sorted_file_count"
echo "Total Folders Created: $sorted_folder_count"
```
