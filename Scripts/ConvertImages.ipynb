# Convert Images

```python
import base64
import pyperclip
from IPython.display import display, Image

# Convert image to base64 string
def image_to_base64(img_path):
    with open(img_path, "rb") as image_file:
        return base64.b64encode(image_file.read()).decode('utf-8')

# Image path
img_path = "C:/Users/samhs/Jupyter(DataBackup)/Images/Details.png"  # replace with your image path

# Get base64 string
img_base64_str = image_to_base64(img_path)

# Display the image in the notebook
img_data = base64.b64decode(img_base64_str)
display(Image(data=img_data))

# Generate the markdown string for the base64 image
markdown_str = f'![Image description](data:image/png;base64,{img_base64_str})'

# Copy the markdown string to clipboard
pyperclip.copy(markdown_str)

print("Image string copied to clipboard!")
```
