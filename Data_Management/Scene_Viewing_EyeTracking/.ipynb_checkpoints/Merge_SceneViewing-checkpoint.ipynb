```python
#!/bin/bash

# Current working directory is the base
base_dir=$(pwd)

# Array of block names
blocks=("Encoding_Block1" "Encoding_Block2" "Encoding_Block3" "Encoding_Block4" \
        "Retrieval_Block1" "Retrieval_Block2" "Retrieval_Block3")

# Create a directory for sorted files
mkdir -p "$base_dir/Sorted"

# Log file for missing sessions and file conflicts
log_file="$base_dir/merge_log.txt"

# Separator line
separator="_________________________________________________________________"

# Append separator to the log file
echo "$separator" >> "$log_file"

# Iterate over each block and then each ID folder within it
for block in "${blocks[@]}"; do
    for id_dir in "$block"/*; do
        if [ -d "$id_dir" ]; then
            id=$(basename "$id_dir")

            # Iterate over session folders within ID folder
            for session_dir in "$id_dir"/*; do
                if [ -d "$session_dir" ]; then
                    session=$(basename "$session_dir")
                    sorted_session_dir="$base_dir/Sorted/$id/$session"

                    # Create corresponding session directory in Sorted
                    mkdir -p "$sorted_session_dir"

                    file_count=0

                    # Move files from this session to the sorted session directory
                    for file in "$session_dir"/*; do
                        if [ -e "$file" ]; then
                            filename=$(basename "$file")
                            dest_file="$sorted_session_dir/$filename"

                            # Check if file already exists in destination
                            if [ -e "$dest_file" ]; then
                                # Log error to log file
                                echo "ERROR: Duplicate file name detected: $filename in session $session for ID $id" >> "$log_file"
                            else
                                mv "$file" "$dest_file"
                                ((file_count++))
                            fi
                        fi
                    done

                    # Output merged files information
                    echo "Merged files from blocks belonging to \"$session\" [*$file_count files in final merged folder]"
                fi
            done
        fi
    done
done

echo "Processing completed. Check $log_file for any issues or details."
```
